```python
import pandapower as pp
net = pp.create_empty_network(sn_mva = 100)
```

```python
bus0 = pp.create_bus(net, name ='Bus0', vn_kv = 22, type = 'b')
bus1 = pp.create_bus(net, name= 'Bus1', vn_kv = 22, type = 'b')
bus2 = pp.create_bus(net, name = 'Bus2', vn_kv = 11, type = 'b')
bus3 = pp.create_bus(net, name = 'Bus3',vn_kv = 220, type = 'b')
bus4 = pp.create_bus(net, name = 'Bus4',vn_kv = 220, type = 'b')
bus5 = pp.create_bus(net, name = 'Bus5',vn_kv = 110, type = 'b')
bus6 = pp.create_bus(net, name = 'Bus6',vn_kv = 110, type = 'b')

net.bus
```

```python
pp.create_ext_grid(net, bus0, va_degree = 0, vm_pu = 1)
```

```python
t1 = pp.create_transformer_from_parameters(net,bus3,bus1,name = 'T1',sn_mva = 50,vn_hv_kv = 220, vn_lv_kv = 22,vk_percent = 10,pfe_kw = 15,vkr_percent = 0,i0_percent = 0.1)
t2 = pp.create_transformer_from_parameters(net,bus4,bus2,name = 'T2',sn_mva = 40,vn_hv_kv = 220, vn_lv_kv = 11,vk_percent = 0.06,pfe_kw = 10,vkr_percent = 0,i0_percent = 0.1)
t3 = pp.create_transformer_from_parameters(net,bus5,bus1,name = 'T3',sn_mva = 40,vn_hv_kv = 110, vn_lv_kv = 22,vk_percent = 0.064,pfe_kw = 10,vkr_percent = 0,i0_percent = 0.1)
t4 = pp.create_transformer_from_parameters(net,bus6,bus2,name = 'T4',sn_mva = 40,vn_hv_kv = 110, vn_lv_kv = 11,vk_percent = 0.08,pfe_kw = 10,vkr_percent = 0,i0_percent = 0.1)
```

```python
net.trafo
```

```python
load = pp.create_load_from_cosphi(net, bus2, scaling = 0.9, sn_mva = 57,const_z_percent = 100, cos_phi= 0.6,mode='underexcited')
net.load
```

```python
z = pp.create_impedance(net, bus0, bus1, r_pu =0, x_pu = 0.18,sn_mva = 90,rft_pu=0,xft_pu=0.18 )
```

```python
l1 = pp.create_line_from_parameters(net,bus3,bus4,length_km=1,r_ohm_per_km=0,x_ohm_per_km=48.4,c_nf_per_km=0,max_i_ka = 50)
l2 = pp.create_line_from_parameters(net,bus5,bus6,length_km=1,r_ohm_per_km=0,x_ohm_per_km=65.43,c_nf_per_km=0,max_i_ka = 50)
```

```python
pp.runpp(net,max_iteration = 100)
```
